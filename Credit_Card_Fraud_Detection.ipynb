```python
# Importing the dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading dataset into pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# Printing the first five rows of the data
credit_card_data.head()
```

```python
# The last five rows of the data
credit_card_data.tail()
```

```python
# Getting number of rows and column
credit_card_data.shape
```

```python
# dataset informations
credit_card_data.info()
```

```python
# Checking the number of missing values
credit_card_data.isnull().sum()
```

**Handling the missing values: Dropping Method**

```python
new_credit_card_data = credit_card_data.dropna(how='any')
```

```python
new_credit_card_data.isnull().sum()
```

Getting the distribution of transactions

```python
# Obtaining the distribution of legit transaction and fraudulent transactions
new_credit_card_data['Class'].value_counts()
```

Handling The Highly Imbalanced data Encountered


1.   Normal/Legit transaction --> 0
2.   Fraudulent transaction --> 1

```python
# Separating the data for analysis
legit = new_credit_card_data[new_credit_card_data.Class == 0]
fraud = new_credit_card_data[new_credit_card_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

Getting the statistical measures of the data


```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Comparing the values for both transactions
new_credit_card_data.groupby('Class').mean()
```

**UNDER-SAMPLING**

```python
# Bulding  a sample dataset containing similar distribution of legit/normal transaction and fraudulent transaction

legit_sample = legit.sample(n=25)
```

```python
# Concatenating two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**SPLITTING DATASET**

```python
# Splitting data into features and targets
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting data into training and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3)
print(X.shape, X_train.shape, X_test.shape)
```


MODEL TRAINING: logistic regression model

```python
# Loading the logistic regression model
model = LogisticRegression()
```

```python
# Training the model with dataset
model.fit(X_train, Y_train)
```

**MODEL EVALUATION**

```python
# Accuracy score on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print('Accuracy score on training data : ', training_data_accuracy)
```

```python
# Accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print('Accuracy score on test data : ', test_data_accuracy)
```
